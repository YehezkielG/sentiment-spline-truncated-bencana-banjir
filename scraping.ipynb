```python
from apify_client import ApifyClient

# Initialize the ApifyClient with your API token
api_token = ""
client = ApifyClient(api_token)

# Prepare the Actor input
run_input = {
    "commentsPerPost": 10000,
    "excludePinnedPosts": False,
    "maxRepliesPerComment": 0,
    "postURLs": [
        "https://www.tiktok.com/@antaranews/video/7580280925297184007",
    ],
    "resultsPerPage": 100
}

# Run the Actor and wait for it to finish
run = client.actor("BDec00yAmCm1QbMEI").call(run_input=run_input)

# Fetch and print Actor results from the run's dataset (if there are any)
for item in client.dataset(run["defaultDatasetId"]).iterate_items():
    print(item)
```

```python
import pandas as pd

data = list( client.dataset(run["defaultDatasetId"]).iterate_items())
df = pd.DataFrame(data)
df.to_csv('data/data unmerge/tiktok_comments5.csv', index=False)
df
```
